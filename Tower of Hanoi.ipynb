```python
def Create_structure(n):
    
    A = []
    B = []
    C = []
    
    for num in range(n):
        
        A.append('  '+' '*(n-num-1)*2+'|'+(num+1)*'‾‾‾‾'+'‾|'+' '*(n-num-1)*2)
        B.append(pattern)
        C.append(pattern)
        
    return A,B,C
```

```python
def Print_towers(A,B,C):
    
    global index
    print(f'\n\n-------------\nSTEP {index}:\n')
    index += 1
    
    for pattern in ([' A ',' B ',' C '],['|‾|']*3,['| |']*3):
        print(' '*(n*2+2)+pattern[0]+' '*(n*4+2)+pattern[1]+' '*(n*4+2)+pattern[2])
    
    for num in range(n):
        print(A[num]+B[num]+C[num])
        
    print('/'+'‾'*(15+n*12)+'\ ')
```

```python
def List_replace(source,dest):
    
    index = 0
    while source[index] == pattern:
        index += 1
    
    dest.pop(0)
    dest.insert(0,source.pop(index))
    source.insert(0,pattern)
     
    index = 0
    while index < n:
        if dest[index] != pattern:
            index += 1  
        else:
            dest.insert(0,dest.pop(index))   
```

```python
def Hanoi_recursion(n,source,mid,dest):
    
    if n > 0:
        Hanoi_recursion(n-1,source,dest,mid)
        Print_towers(A,B,C)
        List_replace(source,dest)  
        Hanoi_recursion(n-1,mid,source,dest)
```

```python
def Input_validation():

    while True:
        n = input("Enter the number of disks (floors) forming the Hanoi Tower (between 1 and 8): ")
        if n.isdigit():
            n = int(n)
            if 0 < n < 9:
                break
            else:
                print("Wrong number, numbers higher than 8 will result in (2^n-1) moves or more.")
                continue     
        print("Invalid Input")
    print("\n")
    return n
```
