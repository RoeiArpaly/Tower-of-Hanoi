## Tower of Hanoi

```python
def Create_structure(n):
    
    tower_one = []
    tower_two = []
    tower_three = []
    
    for num in range(n):

        tower_one.append(f' '+' '*(n-num-1)*2+(num+1)*'____'+' '+' '*(n-num-1)*2)
        tower_one.append(f' '*(n-num-1)*2+'|'+(num+1)*'____'+'|'+' '*(n-num-1)*2)
        
        tower_two.append(' '*(n*4+2))
        tower_two.append(' '*(n*4+2))
        
        tower_three.append(' '*(n*4+2))
        tower_three.append(' '*(n*4+2))
        
    return tower_one,tower_two,tower_three
```

```python
def Print_Towers(a,b,c):
    
    global index
    index += 1
    print(f'\n----------\nSTEP {index}:')
    
    for num in range(len(tower_one)):
        
        print(a[num]+' '*10+b[num]+' '*10+c[num])
        
    print(' '+'_'*(n*2-1)+'||'+'_'*(n*4+10)+'||'+'_'*(n*4+10)+'||'+'_'*(n*2-1)+' ')
    print('|'+' '*(n*2-2)+'TOWER A!'+' '*(n*4+2)+'TOWER B!'+' '*(n*4+2)+'TOWER C!'+' '*(n*2-2)+'|')
    print(' '+''*(24+n*12)+' ')
```

```python
def List_replace(source,dest):
    
    d = {'a':tower_one,'b':tower_two,'c':tower_three}
    
    index = 0
    while d[source][index] == ' '*(n*4+2) and index < len(d[source]):
        index += 1
    
    d[dest].pop(0)
    d[dest].pop(0)
    d[dest].insert(0,d[source].pop(index))
    d[dest].insert(1,d[source].pop(index))
    d[source].insert(0,' '*(n*4+2))
    d[source].insert(0,' '*(n*4+2))
     
    index = 0   
    while index < len(d[source]):
        
        if d[dest][index] != ' '*(n*4+2):
            index += 1  
        else:
            d[dest].insert(0,d[dest].pop(index))   
```

```python
def Hanoi_recursion(n,source,mid,dest):
    
    if n > 0:
        Hanoi_recursion(n-1,source,dest,mid)
        Print_Towers(tower_one,tower_two,tower_three)
        List_replace(source,dest)  
        Hanoi_recursion(n-1,mid,source,dest)
```

```python
def Input_validation():
    
    while True:
        n = input("Enter the number of disks (floors) forming the Hanoi Tower (between 1 and 8): ")
        
        if n.isdigit():
            n = int(n)
            
            if 0 < n < 9:
                break
            else:
                print("Wrong number, numbers higher than 8 will result in 512 moves or more.")
                continue
                
        print("Invalid Input")
    
    print("\n")
    return n
```

```python
if __name__ == "__main__":

    print("The Tower of Hanoi is a mathematical puzzle.\n")
    print("This Algorithm display each step towards the solution (using recursion).\n")

    n = Input_validation()

    structure_list = Create_structure(n)

    tower_one = structure_list[0]
    tower_two = structure_list[1]
    tower_three = structure_list[2]

    index = 0

    Hanoi_recursion(n,'a','b','c')

    Print_Towers(tower_one,tower_two,tower_three)

    print("\n----------\n- END -")
```
